```python
#log to lin uncertainty
import numpy as np

def linear_to_log_uncertainty(lin_x, lin_delta_x):
    return lin_delta_x / (lin_x * np.log(10))

def log_to_linear_uncertainty(log_x, log_x_uncertainty):
    return float(log_x_uncertainty) * 10**float(log_x) * np.log(10)
```

```python
#mass and distance -> BHS for all sources

import numpy as np

def getBHS(logmass, logmasserr, dist, disterr):
    mass=10**logmass
    mass=mass*1.989*10**(30)
    dist=dist*10**6
    dist=dist*30856775814913673
    R_S=2*G*mass/c**2
    bhs=np.sqrt(27)*R_S/dist
    bhs=bhs*360/(2*np.pi)
    bhs=bhs*3600
    bhs=bhs*10**6
    if logmasserr == 'NA':
        return(bhs)
    #log_to_linear_uncertainty(log_x, log_x_uncertainty)
    disterr=disterr*10**6
    disterr=disterr*30856775814913673
    masserr=log_to_linear_uncertainty(logmass, logmasserr)
    masserr=masserr*1.989*10**(30)
    R_Serr=2*G*masserr/c**2
    bhserr=np.sqrt((np.sqrt(27)*R_Serr/dist)**2+(np.sqrt(27)*R_S*disterr/dist**2)**2)
    bhserr=bhserr*360/(2*np.pi)
    bhserr=bhserr*3600
    bhserr=bhserr*10**6
    return(bhs,bhserr)

hubble=73.9
hubbleerr=3.0

G=6.67430*10**(-11)
c=299792458

val=466.87
valerr=0.49
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.58
logmasserr=0.03
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 4258 & {bhs:.3f} $\pm$ {bhserr:.3f}")

val=4088.8
valerr=5.3
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.85
logmasserr=0.05
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC1194 & {bhs:.3f} $\pm$ {bhserr:.3f}")

val=1137
valerr=3
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
#dist=13.97
#disterr=2.1
#logmass=6.92
#logmasserr=0.25
mass=16.6*10**6
masserr=0.1*10**6
logmass=np.log10(mass)
logmasserr=linear_to_log_uncertainty(mass, masserr)
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
#print()
#print(f"{logmass:.2f}")
#print(f"{logmasserr:.3f}")
print(f"NGC 1068 & {bhs:.3f} $\pm$ {bhserr:.3f}")
#print()

val=8322.22
valerr=1.13
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.64
logmasserr=0.05
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 5765b & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=1850.8
valerr1=13.5
valerr2=13.9
dist=val/hubble
disterr1=np.sqrt((valerr1/hubble)**2+(hubbleerr*val/hubble**2)**2)
disterr2=np.sqrt((valerr2/hubble)**2+(hubbleerr*val/hubble**2)**2)
disterr=(disterr1+disterr2)/2
logmass=6.93
logmasserr=0.04
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 2273 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=3833
valerr=5
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.20
logmasserr=0.33
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 3393 & {bhs:.4f} $\pm$ {bhserr:.4f}")

mM=28.12
mMerr=0.36
dist=10**((mM+5)/5)/10**6
disterr=np.log(10)/5*dist*mMerr
logmass=6.06
logmasserr=0.10
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"Circinus & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=6915
valerr=3
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.384
logmasserr=0.038
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"CGCG 074-06 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=10189.26
valerr=1.20
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.49
logmasserr=0.05
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 6264 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=3259.75
valerr=1.00
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.99
logmasserr=0.09
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"UGC 3789 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=2524
valerr=1
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.86
logmasserr=0.04
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 4388 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=563
valerr=3
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.14
logmasserr=0.18
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)

print(f"NGC 4945 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=10803
valerr=3
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.41
logmasserr=0.02
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"UGC 6093 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=1106
valerr=1
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.40
logmasserr=0.05
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 3079 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=4954.5
valerr=15
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.03
logmasserr=0.05
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 2960 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=7618.2
valerr=14.0
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.22
logmasserr=0.03
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"ESO 558 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=2783
valerr=10
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.74
logmasserr=0.16
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 1320 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=2925
valerr=2
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.64
logmasserr=0.06
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"IC 2560 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=7834.28
valerr1=2.1
valerr2=2.2
dist=val/hubble
disterr1=np.sqrt((valerr1/hubble)**2+(hubbleerr*val/hubble**2)**2)
disterr2=np.sqrt((valerr2/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.00
logmasserr=0.05
disterr=(disterr1+disterr2)/2
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"NGC 6323 & {bhs:.5f} $\pm$ {bhserr:.5f}")

val=4818
valerr=10.5
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=7.00
logmasserr=0.06
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"J0437+2456 & {bhs:.4f} $\pm$ {bhserr:.4f}")

val=9076
valerr=32
dist=val/hubble
disterr=np.sqrt((valerr/hubble)**2+(hubbleerr*val/hubble**2)**2)
logmass=6.28
logmasserr=0.13
bhs,bhserr=getBHS(logmass, logmasserr, dist, disterr)
print(f"Mrk 1029 & {bhs:.5f} $\pm$ {bhserr:.5f}")
```
