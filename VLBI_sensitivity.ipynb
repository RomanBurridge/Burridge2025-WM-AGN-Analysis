```python
#raw sensitivity
import numpy as np

def getsnr(cht):
    sefd1=4000
    sefd2=74
    bndw=4*10**9
    nsys=0.88

    nom=4000*74
    denom=2*cht*bndw
    noiseJy=1/nsys*np.sqrt(nom/denom)
    noisemJy=noiseJy*1000

    signalmJy=10
    snr=signalmJy/noisemJy
    return(noisemJy,snr)

chts=[10,1000,10000]
for cht in chts:
    noisemJy,snr=getsnr(cht)
    print(cht)
    print(f"{noisemJy:.2f}")
    print(f"{snr:.1f}")
    print()
```

```python
#astrometric precision
import numpy as np

def getsnr(beam,prec,precunc):
    factor=(4/np.pi)**(1/4)*1/(np.sqrt(8*np.log(2)))
    snr=factor*beam/prec
    snrunc=factor*beam*precunc/(prec**2)
    return(snr,snrunc)

prec=0.086
precunc=0.007
#L2
beam=0.18
print('L2')
print(getsnr(beam,prec,precunc))
print()
#moon
print('moon')
beam=0.69
print(getsnr(beam,prec,precunc))

```
